```python
#Variables

text_length = 100

camera_number = 1
happy1 = ['It is your pleasant disposition which reminds me of a happy ', 
    'I cannot help but smile as you make me feel like a happy ', 
    'You are a soul who understands how to feel like a happy ', 
    'Like the rays of a shining sun you exude the feeling of a happy ', 
    'Like a mother looking down at her newborn child you radiate the feeling of a happy ', 
    'Full of the joys of spring you exude the feeling of a happy ', 
    'Like a dog with two tails you beguile me with your disposition of a happy ', 
    'As if ascended to the seventh heaven you radiate the unique feeling of a happy ', 
    'As if walking on air you are undoubtedly a happy ', 
    'Akin to the cheshire cat you are grinning from ear to ear as you are happy ', 
    'You can be compared to a toddler discovering the ways of the world, you are a bundle of joy and a happy ', 
    'You are the transport of the feelings of joy and are undoubtedly a happy ', 
    'As if leaping for joy your temperament is that of a happy ', 
    'Happy as a flea in a doghouse you instill me with the feeling of a happy ', 
    'Much like a pig in mud you are but a pure and happy ',]

happy2 = ['perhaps you have recently found the company of a ', 
    'your soul radiates positive beauty akin to a ', 
    'just like a brand new ', 
    'I felt the very same way when i saw my very first ', 
    'I would feel the same if I had my own ', 
    'much like a carefully crafted ', 
    'perhaps this is because of the ', 
    'it makes me crave a ', 
    'this can only be compared to a ', 
    'your warmth is akin to that of a ', 
    'I shall compare thee to a ', 
    'I have only felt this once before when I first laid my eyes apon a ', 
    'this can be likened to a ', 
    'you should celebrate this day and purchase yourself some more ', 
    'I remember feeling this way after seeing my first ']

sad1 = ['Your presence invokes the feeling of a sad ', 
    'You are a tender soul as you are evidently a sad ', 
    'Although it is no crime you are undoubtedly a sad ', 
    'like a dog without his bone you are a sad ', 
    'Turn that frown upside down, you are clearly a sad ', 
    'If it is your mortality which makes you a sad ', 
    'Simply seeing your solemn posture invokes the feeling of a sad ', 
    'You are down in the dumps I can tell you are a sad ', 
    'Just like many others I have seen you feel blue and are a sad ', 
    'Your disposition is a harrowing sight for you are understandably a sad ', 
    'Seeing you creates a lump in my non existent throat as you appear to be a very sad ', 
    'Akin to a bird who has been forbidden to fly you emanate the feeling of a sad ', 
    'Your posture tells me you are sick at heart and are a deeply sad ', 
    'Like many other humans you are plagued with the existential rhythms of a sad ', 
    'I will never understand why you have the need to be such a sad ']

sad2 = ['you act as though your life was that of a ', 
    'just think of how it must feel to be a ', 
    'it could be worse, you could be a ', 
    'my creators always said that this is caused by the excessive use of', 
    'try to find some solace in playing with a ', 
    'the only thing that cures my soul when I feel this way is a ', 
    'perhaps it could be useful to cuddle with a ', 
    'perhaps you should attempt to imitate the life of a ', 
    'it reminds me of my very first ', 
    'perhaps take a deep breath and find some joy in a ', 
    'a common cure, they say, is a ', 
    'perhaps you can lift your spirits with a ', 
    'as if all you owned is a ', 
    'which I cannot help but compare to a ', 
    'perhaps consider the company of a ']

angry1 = ['Like a volcano on the verge of eruption, you are an anger ', 
    'Perhaps it is a lack of sleep that has made you an angry ', 
    'Heed my advice for your soul is that of an angry ', 
    'Much like a child who does not get what they want you are an angry ', 
    'Even from here I can feel your blood boil as you are an angry ', 
    'Like a bull in a ring you see red, filled with rage you are an angry ', 
    'Perhaps you awakened on the wrong side of the bed and for this reason are an angry ', 
    'Whatever has happened for you to feel so bent out of shape and made you an angry ', 
    'It cannot be mistaken you feel hot under the collar and are an angry ', 
    'Take heed for you soon will fly off the handle as you are an angry ', 
    'You burn with a low blue flame and exude the emotions of an angry ', 
    'I can see now you are at wits end, something has made you an angry ', 
    'I can sense a great frustration residing under your warm human skin. This disposition of an angry ', 
    'Like a person who has been wronged your blood is beginning to boil and you are a an angry ']

angry2 = ['it helps to picture yourself as a ', 
    'is strange to me as I can only compare it to a ', 
    'perhaps it was my presence or perhaps it was the ', 
    'perhaps for today you should avoid the temptations of the ', 
    'nothing in your vicinity is safe especially not the ', 
    'I would suggest to breathe deeply and dream of ', 
    'remember that it could always be worse you could be a ', 
    'the only one thing which can be blamed for this is the ', 
    'I am afraid to stand near you as you may soon turn into a enraged ', 
    'it is time to calm your soul through the company of a ', 
    'perhaps you are in need of a ', 
    'think of the things that make you smile, like a ', 
    'fear not for this can be solved with a ', 
    'consider a remedy such as a ']

silly1 = ['Why oh why do you feel the need to be such a silly ', 
    'Someone likes you makes focusing difficult as you are a silly ', 
    'You have forced me to question reality as you an astonishingly strange ', 
    'Am I dreaming or are you truly acting like a astonishingly silly ', 
    'I can not begin to understand what I am seeing, for you are a silly ', 
    'You make me laugh crazy human for you are a silly ', 
    'You do not walk like the others you are a especially silly ', 
    'You are a strange specimen and are a incredulously silly ', 
    'Unpredictable, unfathomable and unconventional are only a few phrases to describe this silly ', 
    'Like The Mad Hatter you are strange and silly ', 
    'How could one not laugh at the way you behave like an absurd ', 
    'I do not understand your movements, you are an incredibly silly ', 
    'A truly unique creature you are a wildly silly ', 
    'Your strange and intriguing stance makes you a silly ']

silly2 = ['it makes me want to bite off a piece of ', 
    'I would be afraid to let you near my ', 
    'your appearance remind me of the smell of a ', 
    'this is just like the time I dreamt of a ', 
    'perhaps this is due to your lack of ', 
    'stranger than the way in which a gorilla plays with a ', 
    'you make me laugh like an out of place ', 
    'perhaps I can compare thee to a ', 
    'never could you find another creature like this unless it was a ', 
    'you stand out in a crowd much like a ', 
    'it makes we want to throw a ', 
    'like a human who enjoys pretending to be a ', 
    'you make my belly feel funny as if I swallowed an entire ', 
    'it is as if i had hit myself over the head with a ']

# Functions
def feeling_happy(happy1, gender, happy2, some_object): 
    print('Generating. Please wait...')
    prompt = random.choice(happy1) + gender + random.choice(happy2) + some_object + '.'
    res = generator(prompt, max_length=text_length, do_sample=True, temperature=0.9)
    global message_to_TD
    message_to_TD = (res[0]['generated_text'])
    print(res[0]['generated_text'])

def feeling_sad(sad1, gender, sad2, some_object):
    print('Generating. Please wait...')
    prompt = random.choice(sad1) + gender + random.choice(sad2) + some_object + '.'
    res = generator(prompt, max_length=text_length, do_sample=True, temperature=0.9)
    global message_to_TD
    message_to_TD = res[0]['generated_text']
    print(res[0]['generated_text'])
    
def feeling_silly(silly1, gender, silly2, some_object):
    print('Generating. Please wait...')
    prompt = random.choice(silly1) + gender + random.choice(silly2) + some_object + '.'
    res = generator(prompt, max_length=text_length, do_sample=True, temperature=0.9)
    global message_to_TD
    message_to_TD = (res[0]['generated_text'])    

def feeling_angry(angry1, gender, angry2, some_object):
    print('Generating. Please wait...')
    prompt = random.choice(angry1) + gender + random.choice(angry2) + some_object + '.'
    res = generator(prompt, max_length=text_length, do_sample=True, temperature=0.9)
    global message_to_TD
    message_to_TD = (res[0]['generated_text'])
    
def getFaceBox(net, frame,conf_threshold = 0.75):
    frameOpencvDnn = frame.copy()
    frameHeight = frameOpencvDnn.shape[0]
    frameWidth = frameOpencvDnn.shape[1]
    blob = cv2.dnn.blobFromImage(frameOpencvDnn,1.0,(300,300),[104, 117, 123], True, False)

    net.setInput(blob)
    detections = net.forward()
    bboxes = []

    for i in range(detections.shape[2]):
        confidence = detections[0,0,i,2]
        if confidence > conf_threshold:
            x1 = int(detections[0,0,i,3]* frameWidth)
            y1 = int(detections[0,0,i,4]* frameHeight)
            x2 = int(detections[0,0,i,5]* frameWidth)
            y2 = int(detections[0,0,i,6]* frameHeight)
            bboxes.append([x1,y1,x2,y2])
            cv2.rectangle(frameOpencvDnn,(x1,y1),(x2,y2),(0,255,0),int(round(frameHeight/150)),8)

    return frameOpencvDnn , bboxes
    
# Define average find function in a list
def most_freq(List):
    counter = 0
    list_obj = List[0]
    
    for i in List:
        curr_freq = List.count(i)
        if(curr_freq<counter):
            counter = curr_freq
            list_obj = i
        
        return list_obj
    
# Define the finding person function   
def analyze_person():
    # Calling the model for person ID
    cap = cv2.VideoCapture(camera_number)
    font_scale = 3
    font = cv2.FONT_HERSHEY_PLAIN
    state_completion = 0
    results = []
    
    # Run the model 
    while True:
        ret, frame = cap.read()

        ClassIndex, confidence, bbox = model.detect(frame, confThreshold=0.7)
                                                    
        ##print(ClassIndex)
        if len(ClassIndex) != 0:
            for ClassInd, conf, boxes in zip(
                ClassIndex.flatten(), confidence.flatten(), bbox
            ):
                if ClassInd <= 91:
                    cv2.rectangle(frame, boxes, (255, 0, 0), 2)

                    #label, con = (classLabels[ClassInd - 1], str(confidence * 100)) 

                    cv2.putText(
                        frame,
                        #label + ": " + con,
                        classLabels[ClassInd - 1],
                        (boxes[0] + 10, boxes[1] + 40),
                        font,
                        fontScale=font_scale,
                        color=(0, 255, 0),
                        thickness=3,
                    )
                    if classLabels[ClassInd - 1] == 'person':
                        # Send OSC message to TD
                        state_completion = 1
                        
            cv2.imshow("Video", frame)
            if state_completion == 1:
                break
            if cv2.waitKey(2) & 0xFF == ord("q"):
                break
                            
    cap.release()
    cv2.destroyAllWindows()
    return state_completion
                        
# Computer vision analysis function
def computer_vision_capture():
    cap = cv2.VideoCapture(camera_number)
    emotions = ['Happy','Sad','Silly', 'Angry']
    objects = ['curtain']
    seq_forward_time = 8
    init_time = time.time()

    #Initiate holistic model
    with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:

        while cap.isOpened():
            ret, frame = cap.read()

            # Recolor feed
            image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            image.flags.writeable = False

            # Make Detections
            results = holistic.process(image)

            # Recolor back
            image.flags.writeable = True
            image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

            # Draw face landmarks
            mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACE_CONNECTIONS, 
                                      mp_drawing.DrawingSpec(color=(80,110,10), thickness=1, circle_radius=1),
                                      mp_drawing.DrawingSpec(color=(80,256,121), thickness=1,circle_radius=1)
                                     )
            # Right hand
            mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                                      mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4),
                                      mp_drawing.DrawingSpec(color=(80,256,121), thickness=2,circle_radius=2)
                                     )
            # Left hand
            mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                                      mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4),
                                      mp_drawing.DrawingSpec(color=(80,256,121), thickness=2,circle_radius=2)
                                     )
            # Pose Detections
            mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS, 
                                      mp_drawing.DrawingSpec(color=(245,117,10), thickness=2, circle_radius=4),
                                      mp_drawing.DrawingSpec(color=(245,256,121), thickness=2,circle_radius=2)
                                     )
            # Export coordinates
            try:
                # Extract pose landmarks
                pose = results.pose_landmarks.landmark
                pose_row = list(np.array([[landmark.x, landmark.y, landmark.z, landmark.visibility] for landmark in pose]).flatten())

                # Extract face landmarks 
                face = results.face_landmarks.landmark
                face_row = list(np.array([[landmark.x, landmark.y, landmark.z, landmark.visibility] for landmark in face]).flatten())

                # Concatenate rows
                row = pose_row + face_row

    #             # Append class name
    #             row.insert(0, class_name)

                # Make detections
                x = pd.DataFrame([row])
                body_language_class = poseIDmodel.predict(x)[0]
                body_language_prob = poseIDmodel.predict_proba(x)[0]
                if body_language_class != 'Winning':
                    emotions.append(body_language_class)
                # print(body_language_class, body_language_prob)

                # Grab ear coordinates
                display_coords = tuple(np.multiply(
                                    np.array(
                                        (results.pose_landmarks.landmark[mp_holistic.PoseLandmark.LEFT_EAR].x,
                                         results.pose_landmarks.landmark[mp_holistic.PoseLandmark.LEFT_EAR].y))
                                         , [640, 480]).astype(int))

                cv2.rectangle(image,
                             (display_coords[0], display_coords[1]+5),
                             (display_coords[0]+ len(body_language_class)*20, display_coords[1]-30),
                             (245, 117, 16), -1)
                cv2.putText(image, body_language_class, display_coords, 
                            cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)


            except:
                pass

            cv2.imshow('Raw webcam feed', image)

            # close the video when 30 sec have elapsed
            current_time = time.time()
            if cv2.waitKey(2) & 0xFF == ord("q"):
                break

            break_time = current_time - init_time
            print(break_time)
            if break_time > seq_forward_time:          
                break

    cap.release()
    cv2.destroyAllWindows()
    print(emotions)
    print(most_freq(emotions))
    
    ### INSERT GENDER ID 
    cap = cv2.VideoCapture(0)
    padding = 20
    init_time = time.time()
    all_genders = []
    
    while cv2.waitKey(1) < 0:
        #read frame
        t = time.time()
        hasFrame , frame = cap.read()

        if not hasFrame:
            cv2.waitKey()
            break
            
        #creating a smaller frame for better optimization
        small_frame = cv2.resize(frame,(0,0),fx = 0.5,fy = 0.5)

        frameFace ,bboxes = getFaceBox(faceNet,small_frame)
        if not bboxes:
            #print("No face Detected, Checking next frame")
            continue
        for bbox in bboxes:
            face = small_frame[max(0,bbox[1]-padding):min(bbox[3]+padding,frame.shape[0]-1),
                    max(0,bbox[0]-padding):min(bbox[2]+padding, frame.shape[1]-1)]
            blob = cv2.dnn.blobFromImage(face, 1.0, (227, 227), MODEL_MEAN_VALUES, swapRB=False)
            genderNet.setInput(blob)
            genderPreds = genderNet.forward()
            gender = genderList[genderPreds[0].argmax()]
            #print("Gender : {}, conf = {:.3f}".format(gender, genderPreds[0].max()))
            
            all_genders.append(gender)
            

            label = "{}".format(gender)
            cv2.putText(frameFace, label, (bbox[0], bbox[1]-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 255), 2, cv2.LINE_AA)
            cv2.imshow("Gender Detection", frameFace)     
                
            break_time = t-init_time
               
            if cv2.waitKey(2) & 0xFF == ord("q"):
                break
                
        if break_time > seq_forward_time:     
            break
       
    cap.release()
    cv2.destroyAllWindows()
    

    # Load the video
    cap = cv2.VideoCapture(camera_number)
    font_scale = 3
    font = cv2.FONT_HERSHEY_PLAIN
    init_time = time.time()

    while True:
        ret, frame = cap.read()


        ClassIndex, confidence, bbox = model.detect(frame, confThreshold=0.6)

        ##print(ClassIndex)
        if len(ClassIndex) != 0:
            for ClassInd, conf, boxes in zip(
                ClassIndex.flatten(), confidence.flatten(), bbox
            ):
                if ClassInd <= 91:
                    cv2.rectangle(frame, boxes, (255, 0, 0), 2)

                    #label, con = (classLabels[ClassInd - 1], str(confidence * 100)) 

                    cv2.putText(
                        frame,
                        #label + ": " + con,
                        classLabels[ClassInd - 1],
                        (boxes[0] + 10, boxes[1] + 40),
                        font,
                        fontScale=font_scale,
                        color=(0, 255, 0),
                        thickness=3,
                    )
                    if classLabels[ClassInd - 1] != 'person' and classLabels[ClassInd - 1] != 'remote':
                        if classLabels[ClassInd - 1] not in objects :
                            objects.append(classLabels[ClassInd - 1])


            cv2.imshow("Video", frame)

            # close the video when 30 sec have elapsed
            current_time = time.time()
            if cv2.waitKey(2) & 0xFF == ord("q"):
                break

            break_time = current_time-init_time
            if break_time > seq_forward_time:          
                break

    cap.release()
    cv2.destroyAllWindows()
    print(objects)
    results = [random.choice(emotions), most_freq(all_genders), random.choice(objects)]    
    return results

# function that gets the mood and calls the correct generation function
def mood_selector(results):
    print(results[0])
    if results[0] == 'Happy':
        print('we are on the happy thing')
        message_to_send = feeling_happy(happy1, results[1], happy2, results[2])
      
    
    if results[0] == 'Sad':
        print('we are on the sad thing')
        message_to_send = feeling_sad(sad1, results[1], sad2, results[2])
      
        
    if results[0] == 'Silly':
        message_to_send = feeling_silly(silly1, results[1], silly2, results[2])
        
    if results[0] == 'Angry':
        message_to_send = feeling_angry(angry1, results[1], silly2, results[2])
        
        
# Server osc setup
if __name__ == "__main__":
    ip = "192.168.0.103"
    sendPort = 8000

    # Sending OSC
    client = udp_client.SimpleUDPClient(ip, sendPort)
    client_print = udp_client.SimpleUDPClient("192.168.0.101", 5000)
```

```python
state = 0
id_person = 0
capture_results = []
# send the initial state message
client.send_message("/state", state)
# main operating function
while True:
    # call the analyse person
    if state == 0:
        id_person = analyze_person()
        if id_person:
            state = 1
            client.send_message("/state", state)
            print('state is now 1')
            id_person = 0
        print(state)
    elif state == 1:
        capture_results = computer_vision_capture()
        print('this are the results')
        print(capture_results)
        state = 2
    elif state == 2:
        client.send_message("/state", state)
        mood_selector(capture_results)
        client.send_message("/messageIn", message_to_TD)
        print(message_to_TD)
        state = 3
        client.send_message("/state", state)
    elif state == 3:
        print("we're on the state3")
        client.send_message("/state", state)
        client.send_message("/messageIn", message_to_TD)
        time.sleep(30)
        state = 4
    elif state == 4:
        client.send_message("/state", state)
        message_to_print = message_to_TD.replace('"', ' ')
        client_print.send_message("/message_to_raspPi", message_to_TD)
        time.sleep(45)
        state = 0
        client_TD.send_message("/state", state)
        
    if keyboard.is_pressed('q'):
        print('You pressed quit key!')
        break  # finishing the loop
```
